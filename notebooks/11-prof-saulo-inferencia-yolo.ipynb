```python
import os
import sys

# --- TRAVA DE SEGURANÇA MÁXIMA ---
# Força o notebook a ler as bibliotecas da sua .venv, mesmo estando no Python global
caminho_venv = os.path.abspath(os.path.join(os.getcwd(), '..', '.venv', 'Lib', 'site-packages'))
if os.path.exists(caminho_venv) and caminho_venv not in sys.path:
    sys.path.insert(0, caminho_venv)

from ultralytics import YOLO
import torch

# 1. Carrega os seus pesos customizados da bola de tênis
model = YOLO('modelo_bolatenis.pt')

# 2. Define o caminho do seu vídeo
source = '../tenis.mp4'

if os.path.exists(source):
    print(f"✅ Vídeo real encontrado em: {source}")
    
    # Define se vai usar GPU ou CPU de forma segura
    device_atual = 0 if torch.cuda.is_available() else 'cpu'
    print(f"🚀 Iniciando processamento usando: {device_atual}")
    
    # Executa a previsão frame por frame
    results_stream = model.predict(
        source=source, 
        save=True,          
        conf=0.35,          
        imgsz=640,          
        stream=True,        
        device=device_atual # Usa a RTX se o import funcionou, ou CPU para não travar
    )
    
    # Consome o gerador para processar até o final
    for r in results_stream:
        pass
        
    print("\n🎉 Seu vídeo foi processado com sucesso absoluto!")
    print("Assista ao resultado abrindo a pasta: notebooks/runs/detect/predict/")
else:
    print(f"❌ O arquivo não foi encontrado no caminho: {source}")
```
