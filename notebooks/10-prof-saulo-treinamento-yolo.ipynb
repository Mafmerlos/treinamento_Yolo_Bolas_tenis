```python
import os
import shutil
import torch
from ultralytics import YOLO, settings
from IPython.display import Image, display

# 1. Configura a pasta de saída para os treinos ficarem organizados localmente
settings.update({'runs_dir': './runs'})

# 2. Validação crucial que fizemos para garantir o uso da sua placa
print("--- VERIFICAÇÃO DE HARDWARE ---")
print(f"CUDA está ativo e disponível? {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU Detectada: {torch.cuda.get_device_name(0)} (Pronto para o treino!)")
else:
    print("⚠️ ALERTA: GPU não detectada. O treino rodará na CPU e será mais lento.")
```

```python
# Definindo o caminho para o seu arquivo de configuração atual do tênis
dataset_path = '../dataset_bolatenis/data.yaml'

if not os.path.exists(dataset_path):
    raise FileNotFoundError(f"❌ Arquivo '{dataset_path}' não encontrado. Verifique se a pasta 'dataset_bolatenis' está no nível correto da árvore de arquivos.")
else:
    print(f"✅ Dataset de tênis mapeado com sucesso: {dataset_path}")
```

```python
# Carrega a arquitetura base do YOLOv8 Nano pré-treinada
model = YOLO('yolov8n.pt')

# Inicia o treinamento completo na sua RTX 3050
# Definido workers=0 para garantir estabilidade absoluta no ecossistema Windows
results = model.train(
    data=dataset_path,    # '../dataset_bolatenis/data.yaml'
    epochs=100,           # 100 épocas para estabilizar as métricas de perda (loss) da bola
    imgsz=640,            # Resolução cheia de 640x640 (excelente para objetos pequenos e rápidos)
    batch=16,             # Processa 16 imagens por lote (ótimo aproveitamento dos 6GB de VRAM)
    workers=0,            # Evita conflitos de multiprocessing e congelamento do Kernel no Windows
    cache=False,          # Desativado para balancear o consumo de memória RAM do seu sistema
    device=0,             # Força o uso da sua placa de vídeo NVIDIA dedicada
    save=True,            # Salva o melhor modelo em weights/best.pt automaticamente
    plots=True            # Gera todas as matrizes e curvas de aprendizado
)

print("🚀 Treinamento finalizado! Os pesos e métricas foram salvos com sucesso.")
```
